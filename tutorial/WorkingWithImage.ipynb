# Working With Images

This section introduces the usage of BboxTools to work with images, which is also the basic goal of this tools. \
Notice matplotlib is not necessary for BboxTools. But in order to make the visualization inside jupyter notebook, this tutorial uses matplotlib to visualize images. \
The example image is from ImageNet dataset, bounding box and keypoints annotations are from PASCAL3D+ dataset.

```python
import BboxTools as bbt
import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
# Load the image and convert to ndarry
img = np.array(Image.open('example.jpg'))
plt.imshow(img)
```

One usage of BboxTools is data preprocess for object detection. The from_numpy allow you to create a bbox from numpy.ndarry, which is most common annotation format. \
Notice: The default sorts of bbox in pillow is ('x0', 'y0', 'x1', 'y1'), while in BboxTools is ('y0', 'y1', 'x0', 'x1')

```python
# Create the bounding box of the object
box_obj = bbt.from_numpy(np.array([ 83,  72, 367, 228]), sorts=('x0', 'y0', 'x1', 'y1'), image_boundary=img.shape)
print(box_obj)
plt.imshow(box_obj.apply(img))
```

```python
# You can also have some padding before doing the cropping
plt.imshow(box_obj.pad(20).apply(img))
```

**Visualizing bbox** \
Function draw_bbox can visualize a bbox. Currently import image could only be numpy.ndarray. \
Notice draw_bbox is a inplace function, which means it will modified the image. If you want to keep the original image, please using img.copy()

```python
# You can visualize the bbox by drawing the boundary
img_get = bbt.draw_bbox(img.copy(), box_obj, boundary=(255, 0, 0), boundary_width=4)
plt.imshow(img_get)
```

```python
# You can also visualize the bbox by transparently filling the bbox
img_get = bbt.draw_bbox(img.copy(), box_obj, fill=(255, 0, 0, 0.5))
plt.imshow(img_get)
```

**Working with keypoints** \
You can also visualize keypoints with BboxTools. 

```python
# Define the colors to visualize different keypoints
colors = [(255, 0, 0),  (240, 240, 0),  (0, 255, 0),  (0, 240, 240), (0, 0, 255), (240, 0, 240)]

# Define the kp_drawing function. 
def draw_kp_list(im, kp_list, kp_colors, point_size=11):
    imd = ImageDraw.ImageDraw(im)
    for c, k in zip(kp_colors, kp_list):
        # Create a bbox center at this keypoint, with size point_size.
        this_bbox = bbt.box_by_shape((point_size, point_size), (int(k[0]), int(k[1])), image_boundary=im.size[::-1])
        
        # bbox.pillow_box() is a function that convert is bbox in pillow format ('x0', 'y0', 'x1', 'y1')
        imd.ellipse(this_bbox.pillow_bbox(), fill=c)
    return
```

```python
# A set of key points
kps = np.array([[199, 257], [168, 358], [ 80, 193], [ 79, 273], [155,  94], [168, 204]])

# Convert the img from numpy.ndarry to PIL.Image
img_ = Image.fromarray(img)

draw_kp_list(img_, kps, colors)
plt.imshow(np.array(img_))
```

In some case, people detecting keypoints only inside the object bounding box. It is simple to use BboxTools to convert kp to the in-box coordinates. \
Notice: the source_box is bbox of the object in the original image, the target_box is bbox of the object in projected image. In this case the target_box is the whole cropped image, but in some other case it could be part of the image (i.e. pad before cropping).

```python
# Crop the image by the object bounding box
img_in_box = box_obj.apply(img)

# Create the project function, a lambda function is returned
project_func = bbt.projection_function_by_boxes(box_obj, bbt.full(img_in_box.shape))

# Project keypoints to the in-box coordinates
kps_in_box = project_func(kps)

# Now draw the new keypoints
img_in_box_ = Image.fromarray(img_in_box)
draw_kp_list(img_in_box_, kps_in_box, colors)
plt.imshow(np.array(img_in_box_))
```
